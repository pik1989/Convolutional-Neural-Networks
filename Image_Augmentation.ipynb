```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import random
from skimage import exposure
from skimage.util import random_noise
from skimage import transform
from cv2 import resize
```

```python
# Mounting the google colab drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
img = mpimg.imread('/content/drive/MyDrive/AI, Data Science & Analytics/cat_or_dog2.jpg')
```

```python
plt.imshow(img)
```

```python
#RESCALING
```

```python
img_rescale = resize(img, (500,350))
```

```python
plt.imshow(img_rescale)
```

```python
img_rescale = resize(img, (70,50))
```

```python
plt.imshow(img_rescale)
```

```python
#FLIPPING
```

```python
horizontal_flip = np.fliplr(img)
```

```python
plt.imshow(horizontal_flip)
```

```python
vertical_flip = np.flipud(img)
```

```python
plt.imshow(vertical_flip)
```

```python
#Rotation
```

```python
from skimage import transform
trans_img = transform.rotate(img, random.uniform(-40,40))
plt.imshow(trans_img)
```

```python
#Adding Noise
```

```python
img_noise = random_noise(img, mode='s&p', clip = True)
plt.imshow(img_noise)
```
